```python
#Creating products DataFrame
import pandas as pd
d = {
    "Product":["Samsung","Nokia","Samsung","Motorola","Nokia","Samsung","Samsung"],
    "Orders":[2,4,3,4,6,7,3]
}
df1 = pd.DataFrame(d)
print(df1)
```

```python
#Creating products DataFrame,applying groupby
import pandas as pd
d = {
      "Product":["Samsung","Nokia","Samsung","Motorola","Nokia","Samsung","Samsung"],
      "Orders":[2,4,3,4,6,7,3]
}
df1 = pd.DataFrame(d)
grouped = df1.groupby(["Product"])
result = grouped.sum()
print(df1)
print()
print(result)
```

```python
#Applying a single function to columns in groups
import pandas as pd
df1 = pd.read_csv("sales5.csv")
d = {
    "Product_Cost":sum
}
cols = ['Date','Product_Name']
grouped = df1.groupby(cols)
result = grouped.agg(d)
print(result)
```

```python
#Applying a single function to columns in groups
import pandas as pd
df1 = pd.read_csv("sales5.csv")
d = {
    'Product_Cost':sum,
    'Product_Name':"count"
}
cols = ['Date','Product_Name']
grouped = df1.groupby(cols)
result = grouped.agg(d)
print(result)
```

```python
#Applying a single function to columns in groups
import pandas as pd
df1 = pd.read_csv("sales5.csv")
d = {
    "Product_Name":[min,max,sum]
}
col = ["Product_Cost"]
grouped = df1.groupby(col)
result = grouped.agg(d)
print(result)
```
